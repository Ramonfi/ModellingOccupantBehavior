```python
df = ThermalComfort.loc[:, IDX[:,['N', 'S'],:,:]]
PMV = []
for (bui, app, room), group in df.groupby(level=[0,1,2], axis=1):
    group = group.stack([0,1,2]).reset_index()
    group = group.assign(comf = group['pmv'].abs().lt(0.5))
    PMV.append(group.groupby((group.comf != group.comf.shift()).cumsum()).agg({'bui':'first', 'app':'first', 'room': 'first', 'Datetime':['first', 'last', RT.utils.aggMINMAX], 'pmv': ['mean'], 'comf':['mean']}).set_axis(['bui', 'app', 'room', 'start','end', 'duration', 'pmv_mean', 'comfortable'], axis=1).where(lambda df: df.comfortable==0).dropna(how='all'))
PMV = pd.concat(PMV).reset_index(drop=True)
PMV.groupby(['bui', 'app', 'room'])['duration'].agg(['sum', 'mean'])
```

```python
def getOccupied(df):
    if hasattr(df.index, 'freq'):
        _freq_ = df.index.freqstr
        _anw = DB.loc[:, IDX[:,:,'WE', 'Anwesenheit']].resample(_freq_).quantile(.5, interpolation='nearest').droplevel([2,3], axis=1)
        for col, group in df.groupby(level=['bui', 'app'], axis=1):
            df[col] = df.loc[:,col].where(_anw[col]>0)
    return df
    
def style_crit(v, props=''):
    return props if v > 0.02 else None

def getComfortGraphs(bui, app, period, occupied =True, which='all', export=False, **kwargs):
    global TOP
    if 'TOP' not in globals():
        TOP = estimateOperativeTemperature()

    Tamb = getData(bui=bui, app=app, value='T_amb', occupied=occupied).droplevel(3,axis=1).groupby(level=1, axis=1).mean().loc[period]     
    Tair = getData(bui=bui, app=app, value='Tair', occupied=occupied).droplevel(3,axis=1).groupby(level=1, axis=1).mean().loc[period]
    rH = getData(bui=bui, app=app, value='Rh', occupied=occupied).droplevel(3,axis=1).groupby(level=1, axis=1).mean().loc[period]
    if occupied:
        Top = getOccupied(TOP.loc[:, IDX[bui,app,:,:]].droplevel(3,axis=1)).groupby(level=1, axis=1).median().loc[period]
        _title = f'{BUID[bui]} | {APPS[app]} (Nutzungszeitraum)'
    else:
        Top = TOP.loc[:, IDX[bui,app,:,:]].droplevel(3,axis=1).groupby(level=1, axis=1).median().loc[period]
        _title = f'{BUID[bui]} | {APPS[app]}'

    kws = dict(kind='hist', title=_title)
    kws.update(**kwargs)

    if which in ['all', 'NA']:
        RT.comf.adaptive_comfort_NA(Tamb, Top, **kws)
        if export:
            RT.toTex('appendix', f'Beispiel_NA')
    if which in ['all', 'EN']:
        RT.comf.adaptive_comfort_EN(Tamb, Top, **kws)
        if export:
            RT.toTex('appendix', f'Beispiel_EN')
    if which in ['all', 'HX']:
        RT.comf.hxdiagramm(Tair, rH, **kws)
        if export:
            RT.toTex('appendix', f'Beispiel_HX')
```
